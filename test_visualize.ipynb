```python
import os
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import load_img, img_to_array

# Load the trained model
model = load_model("Upload/model.h5")

# Labels
labels = {0: 'Healthy', 1: 'Powdery', 2: 'Rust'}

# Image size used during training
image_size = (224, 224)

# Directory containing test images
test_dir = "Dataset"  # Adjust if your folder name is different

# Function to load and predict an image
def predict_and_show(img_path):
    img = load_img(img_path, target_size=image_size)
    img_array = img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0) / 255.0
    prediction = model.predict(img_array)
    pred_label = labels[np.argmax(prediction)]

    # Display
    plt.imshow(img)
    plt.title(f"Predicted: {pred_label}\nConfidence: {np.max(prediction):.2f}")
    plt.axis('off')
    plt.show()

# List some images from each class
sample_images = [
    "E://SMARTPHONEWHEATDISEASEDETECTION\Dataset\Test\Healthy\8e6a823cce9ff40c.jpg",
    "E://SMARTPHONEWHEATDISEASEDETECTION\Dataset\Test\Powdery\9fd12addd42ac28a.jpg",
    "E://SMARTPHONEWHEATDISEASEDETECTION\Dataset\Test\Rust\85acd23a0b659d97.jpg"
]

# Run predictions
for img_path in sample_images:
    if os.path.exists(img_path):
        predict_and_show(img_path)
    else:
        print(f"Image not found: {img_path}")
```
